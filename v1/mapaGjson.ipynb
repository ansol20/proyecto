```python

from shapely.geometry import Point, LineString, Polygon
import folium
import geopandas as gpd
from IPython.display import clear_output, display

all_positions = []
hospitalPoints = []
maximum_hospital_numbers = 0
for n in data:
  if maximum_hospital_numbers < len(n["hospitales"]):
    maximum_hospital_numbers=len(n["hospitales"])
  for hospitales in n["hospitales"]:
    hospitalPoints.append([hospitales["localizacion"],hospitales["nombre"],len(n["hospitales"])])
  all_positions.append([Polygon(n["posicion"]),len(n["hospitales"])]) 
import pandas as pd
#df=pd.read_json(data)
#print(df)
```

```python
m = folium.Map(location=[-0.2195535, -78.4966275], zoom_start=13) #punto inicial
#colors=["#E8E3E6","#EDDEE8","#E0B8D3","#D98CBF","#E085C2","#E052B1","#D9269D","#AD1F7E","#82175E","#8A0F61","#910863","#990066","#5C0A41","#4D193B","#422438","#3D2936"] #15 colores degradantes 

# These values could be seen as dummy edge weights

datajson = gjsoncuadrante["GJsonCuadrantes"]
 
datapanda = pd.DataFrame(data=gjsoncuadrante["cantidad"])
folium.Choropleth(
    geo_data=datajson,
    name="choropleth",
    data=datapanda,
    columns=["cuadrantes", "cantidad"],
    key_on="feature.id",
    fill_color="PuBu",
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name="Cantidad de Hospitales, Centros medicos (%)",
).add_to(m)

folium.LayerControl().add_to(m)

from colour import Color
gray = Color("gray")
colors = list(gray.range_to(Color("purple"),160))
for i in range(0,len(hospitalPoints)): 
  print(colors[hospitalPoints[i][2]*10])
  folium.CircleMarker(hospitalPoints[i][0],
                radius=2,
                popup=hospitalPoints[i][1],
                color=str(colors[hospitalPoints[i][2]*10]),
                fill_color=str(colors[hospitalPoints[i][2]*10]),
               ).add_to(m)
  clear_output(wait=True)
  print("%d" %i)
```
